**Data cleasning**


```python
gross_null=gross.isna().mean().round(4) * 100
gross_null
```

```python
gross1 = gross.dropna(axis = 1, how = 'all')
# Dropping columns where all rows are NaN
```

```python
gross2 = gross1[gross1.Duration != '2016-17']
```

```python
# Dropping the UT as it is not needed for Analysis

gross3 = gross2.T
gross4 = gross3.drop(labels = ['Andaman & Nicobar Islands','Chandigarh','Delhi','Puducherry'])
```

```python
# Mean of the row (% Growth over previous year) for duration 2013-14, 2014-15 and 2015-16

gross4_mean = gross4.iloc[2:,6:10].mean(axis = 1).round(2).sort_values()
gross4_mean
```

**Data Visualisation and Insights Extractions**

```python
# Bar Plot for Average growth rates of the various states for duration 2013-14, 2014-15 and 2015-16
plt.rcParams['figure.figsize'] = [11.5,8]
colors = ["blue" if i == 'All_India GDP' else "cyan" for i in gross4_mean.index]
gross4_mean.plot(kind='barh',stacked=True, color=colors)
plt.title("Avg.% Growth of States for Duration 2013-14, 2014-15 and 2015-16", fontweight = 'bold')
plt.xlabel("Avg. % Growth", fontweight = 'bold')
plt.ylabel("States", fontweight = 'bold')
```

```python
# Average growth rate of my home state against the National average Growth rate

gross4_myhome = gross4_mean[['Andhra Pradesh ', 'All_India GDP']]
```

```python
gross4_myhome.plot(kind='bar',stacked=True, colormap = 'Dark2')
plt.title("Avg. % Growth of Home State vs National Avg. for Duration 2013-14, 2014-15 and 2015-16", fontweight = 'bold')
plt.ylabel("Average % Growth", fontweight = 'bold')
plt.xlabel("Home State Vs National Average", fontweight = 'bold')
plt.xticks(rotation=0)
```

**Insights from the above plot**

When compared with prior periods, GDP tells us whether the economy is expanding by producing more goods and services, or contracting due to less output. This implies that when my home state Andhra exceeds the national GDP ( All_India GDP) it means that its performing above average over a specific period of time.

**Before the Partition of Andhra and Telengana**

```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
Telangana = gross['Telangana'].iloc[:2]
Andhra_Pradesh = gross['Andhra Pradesh '].iloc[:2]


# Set position of bar on X axis
br1 = np.arange(len(Telangana))
br2 = [x + barWidth for x in br1]


# Make the plot
plt.bar(br1, Telangana, color ='c', width = barWidth,
        edgecolor ='grey', label ='Telangana')
plt.bar(br2, Andhra_Pradesh, color ='b', width = barWidth,
        edgecolor ='grey', label ='Andhra Pradesh')


# Adding Xticks
plt.xlabel('Years', fontweight ='bold', fontsize = 15)
plt.ylabel('GDSP', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Telangana))],
        ['2011-12', '2012-13'])
plt.title('Before partition of Andhra and Telangana')

plt.legend()
plt.show()
```

**After the Partition of Andhra and Telengana**

```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
Telangana = gross['Telangana'].iloc[2:5]
Andhra_Pradesh = gross['Andhra Pradesh '].iloc[2:5]


# Set position of bar on X axis
br1 = np.arange(len(Telangana))
br2 = [x + barWidth for x in br1]


# Make the plot
plt.bar(br1, Telangana, color ='r', width = barWidth,
        edgecolor ='grey', label ='Telangana')
plt.bar(br2, Andhra_Pradesh, color ='g', width = barWidth,
        edgecolor ='grey', label ='Andhra Pradesh')


# Adding Xticks
plt.xlabel('Years', fontweight ='bold', fontsize = 15)
plt.ylabel('GDSP', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Telangana))],
        ['2013-14', '2014-15', '2015-16'])
plt.title('After partition of Andhra and Telengana')

plt.legend()
plt.show()
```

**Comparison of the Average % growth of GDSP/GDP of Telengana, Andhra Pradesh and All India GDP respectfully**

```python
data = gross4_mean[['Telangana', 'Andhra Pradesh ', 'All_India GDP']]
total = sum(data)

explode = (0, 0, 0.05)
NAME = ['Telangana', 'Andhra Pradesh', 'All India GDP']
VALUES = [round(i,2) for i in (list(data))]
plt.pie( data , explode = explode, labels = VALUES)
circle = plt.Circle( (0,0), 0, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
plt.title('Comparison of Average % Growth of GDSP of Telangana, Andhra Pradesh and All India GDP')
plt.legend(NAME)
plt.show()
```

**Comparison of the two highest GDP contributors with the All India GDP of the selected years**
